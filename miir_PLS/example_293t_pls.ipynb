```r
X <- read.csv("../data/final_293t/subsampled_splines.csv", row.names=1)
```

```r
Y <- read.csv("../data/final_293t/PLS/PLS_Y.csv", row.names=1)
```

```r
source("miirPLS.R")
```

```r
#retList = miirPLS(X,Y,n,ModifiedTtest = 1)
options(warn=-1)
retList = miirPLS(X,Y,ModifiedTtest = 1)
options(warn=0)
```

```r
#save.image(file='../data/HaCat_hihghly_var_PLS_workspace.RData')
```

```r
source("miirPLS.R")
options(warn=-1)
#n resamples 
n = 500
#dX number of variables
dX = ncol(X)
#alpha significance level 
alpha = 0.05
#nCut cutoff corresponding to alpha 
nCut = ceiling(dX*alpha)
R <- matrix(0, dX, n)
rCut <- vector("numeric", length = n)
for (j in 1:n){
    #permute X
    Yperm <- Y[sample(nrow(Y)),]
    resampleRetList = miirPLS(X, Yperm, ModifiedTtest = 1)
    for (i in 1:dX){
        #compute R in PLS-loading plane
        R[i,j] = sqrt(sum(resampleRetList$model$sigLoadings[i,]^2))
    }
    r = sort(R[,j], decreasing = TRUE)
    rCut[j] = r[nCut]
}
options(warn=0)
```

```r
mean(rCut)
```

```r
write.csv(retList$q, file="../data/final_293t/PLS/q.csv", row.names=TRUE)
write.csv(retList$model$sigLoadings, file="../data/final_293t/PLS/model_sigloadings.csv", row.names=TRUE)
write.csv(retList$model$scores, file="../data/final_293t/PLS/model_scores.csv", row.names=TRUE)
write.csv(retList$model$loadings, file="../data/final_293t/PLS/model_loadings.csv", row.names=TRUE)
write.csv(retList$model$loadingWeights, file="../data/final_293t/PLS/model_loadingweights.csv", row.names=TRUE)
write.csv(retList$model$sigLoadings, file="../data/final_293t/PLS/model_sigloadings.csv", row.names=TRUE)
write.csv(rCut, file="../data/final_293t/PLS/r_cutoff.csv")
```
